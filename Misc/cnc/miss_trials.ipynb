```python
import numpy as np
import os
import IPython
import soundfile as sf
```

```python
lstm_dir = '/home/yangwenhao/local/project/lstm_speaker_verification'
sv_dir = '/home/yangwenhao/local/project/SpeakerVerification-pytorch'

scores_dir = 'Data/xvector/ThinResNet18/cnceleb/klfb_egs_baseline/arcdist_sgd_rop/Mean_batch256_basic_downk3_none1_SAP2_dp01_alpha0_em512_wd5e4_var/cnceleb_dev_fix'

```

```python
trials = ['data/cnceleb/klfb/dev_fb40/trials_250w', 'data/cnceleb/klfb/dev_fb40/trials_250w2']
scores = ['score.2022.05.07.14:59:58', 'score.2022.05.07.15:03:28']
thresholds = [0.1817, 0.1821]
```

```python
miss_uids = []
numofFN = 0
numofFP = 0

for i,(trial, score) in enumerate(zip(trials, scores)):
    threshold = thresholds[i]
    
    trial_lines = []
    with open(os.path.join(lstm_dir, trial), 'r') as f:
        for l in f.readlines():
            trial_lines.append(l.split())
    
    i = 0
    with open(os.path.join(sv_dir, scores_dir, score), 'r') as f:
        for l in f.readlines():
            truth, s = l.split()
            s = float(s)
            if (s >= threshold and truth == 'True') or (s <= threshold and truth == 'False'):
                pass
            else:
                pairs = trial_lines[i][:2]
                pairs.sort()
                miss_uids.append('-'.join(pairs))
                
                if (s >= threshold and truth == 'False'):
                    numofFP += 1
                elif (s < threshold and truth == 'True'):
                    numofFN += 1
                
            i += 1
            
miss_uids.sort()
```

```python
print(len(miss_uids), numofFP, numofFN)
```
