```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import RidgeCV
import matplotlib.pyplot as plt
```

```python
X, y = load_diabetes(return_X_y=True)
clf = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(X, y)
clf.score(X, y)
```
