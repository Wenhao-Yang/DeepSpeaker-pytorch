```python
def str_hash(name):
    md5 = hashlib.md5()
    md5.update(name.encode('utf-8'))
    
    return md5.hexdigest()
```

```python
acc_files = glob.glob('/home/yangwenhao/project/vits/data/lmp/*.aac')
```

```python
for acc_file in tqdm(acc_files, ncols=50):
# acc_file = '/home/yangwenhao/project/vits/data/lmp/QQ录音 米皮_20220611224327.aac'
    wav_file = acc_file.replace('acc', 'wav')

    song = AudioSegment.from_file(acc_file)
    song.export(wav_file, format='wav')

    samples, sr = sf.read(wav_file)

    file_name = wav_file.split('/')[-1].rstrip('.wav')
    new_name = str_hash(file_name)
    solo_name = '/home/yangwenhao/project/vits/data/lmp2/' + new_name + '.wav'

    sf.write(solo_name, samples[:, 0], sr, subtype='PCM_16')
```
