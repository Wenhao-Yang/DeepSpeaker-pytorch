```python
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np
from scipy.signal import stft
```

```python
ws = ['audio/id10001-1zcIwhmdeo4-00001-nbfm-fading0.wav',
       'audio/id10001-1zcIwhmdeo4-00001-nbfm-fading25e6.wav',
       'audio/id10001-1zcIwhmdeo4-00001-nbfm-fading25e5.wav']
```

```python
plt.figure(figsize=(16,6))

l = len(ws)
for i in range(l):
    # plt.subplot(1,l,i+1)
    wav, sr = sf.read(ws[i])
    
    f,t,zxx = stft(wav, fs=sr, nperseg=0.025*sr, noverlap=0.01*sr, nfft=512, window=np.hamming(0.025*sr))
    abs_zxx = np.log10(np.absolute(zxx))

    # plt.pcolormesh(t, f, abs_zxx)
    # plt.plot(abs_zxx.mean(axis=1))
    print(abs_zxx.shape)
    plt.plot(abs_zxx.mean(axis=1))
    # plt.pcolormesh(t, f, abs_zxx)
plt.legend(['fading-0', 'fading-10/4e6', 'fading-10/4e5', ])
```
