```python
import numpy as np
from scipy.fftpack import fft,ifft
from scipy.signal import stft
import scipy
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# %matplotlib osx
```

```python
fs, wav = wavfile.read('Data/Voxceleb1/vox1_dev_noise/id10001/1zcIwhmdeo4/00001.wav')
f,t,zxx = stft(wav, fs=fs, nperseg=0.025*fs, noverlap=0.01*fs, nfft=512, window=np.hamming(0.025*fs))
abs_zxx = np.absolute(zxx)
print(zxx.dtype)
```

```python
plt.pcolormesh(t, f, abs_zxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
fs, wav = wavfile.read('../data/Nov26_23_11.wav')
```

```python
f,t,zxx = stft(wav[:,0], fs=fs, nperseg=0.025*fs, noverlap=0.01*fs, nfft=2048, window=np.hamming(0.025*fs))
abs_zxx = np.absolute(zxx)
# print(zxx.dtype)
```

```python
print(wav.shape)
print(f.shape)
print(t.shape)
print(zxx.shape)
```

```python
plt.figure(figsize=(16,10))
plt.pcolormesh(t, f[:240], abs_zxx[:240, :])
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
f, t, Sxx = scipy.signal.spectrogram(wav, fs=1.0, window=('tukey', 0.25), nperseg=None, noverlap=None, nfft=None, detrend='constant', return_onesided=True, scaling='density', axis=- 1, mode='psd')
```

```python
# f, t, Sxx = signal.spectrogram(x, fs)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
